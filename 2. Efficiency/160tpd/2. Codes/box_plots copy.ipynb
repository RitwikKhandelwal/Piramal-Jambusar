```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import seaborn
seaborn.set(style='whitegrid')
```

```python
df = pd.read_csv("Flat_File_Forehearth_Level_DPR_Eff_J17.csv",
                 encoding="ISO-8859–1")
```

```python
df["Type"] = ""*len(df)
```

```python
c1 = df.index[(df["DPR Eff"] <= 89)]
c2 = df.index[(df["DPR Eff"] >= 89) &
              (df["DPR Eff"] <= 91)]
c3 = df.index[(df["DPR Eff"] >= 91) &
              (df["DPR Eff"] <= 93)]
c4 = df.index[(df["DPR Eff"] >= 93) &
              (df["DPR Eff"] <= 100)]

c1, c2, c3, c4
```

```python
df.loc[c1, "Type"] = "[0, 89]"
df.loc[c2, "Type"] = "[89, 91]"
df.loc[c3, "Type"] = "[91, 93]"
df.loc[c4, "Type"] = "[93, 100]"
```

## Box Plots

```python
for elem in li:
    if(elem=="Name" or elem=="Date" or elem=="FGCode"):
        continue
    seaborn.boxplot(x="DPR_Eff_Metric",
                y=elem,
                data=df1, order=["[0, 89]", "[89, 91]", "[91, 93]", "[93, 100]"])
    plt.savefig("Box Plots/LineWise/DPR Eff/{}.png".format(elem))
    plt.clf()
```

## Histogram

```python
for elem in li:
    plt.hist(list(df1[elem]), bins=20)
    plt.xlabel(elem)
    plt.savefig("Histogram/{}.png".format(elem))
    plt.clf()
```

## Line Chart


```python
df1 = df1.drop("DPR_Eff_Metric", axis =1)
```

```python
df1["Date"] = pd.to_datetime(df1["Date"]).dt.date
```

```python
df1 = df1.groupby("Date").mean().reset_index()
df1
```

```python
df1["Date"] = pd.to_datetime(df1["Date"], format='%Y-%m-%d')
df1.sort_values(by='Date')
```

```python
df2 = df1[df1["FH 15 ZONE 6 TEMP"] > 1200]

df2
```

```python
chart = seaborn.lineplot(data=df1, x='Date',
                         y='FH 17 ZONE 3 TEMP', palette='Set1')

plt.xticks(rotation=45, horizontalalignment='right', fontweight='light')

locator = mdates.DayLocator(interval=50)
chart.xaxis.set_major_locator(locator)

# plt.show()
plt.savefig(
    "LineChart/{}.png".format('FH 17 ZONE 3 TEMP'), bbox_inches='tight')

# plt.figure(figsize = (16,16))
```

## Variable Importance Plot

```python
df= pd.read_csv("Variable Importance.csv", header = [0])
```

```python
df
```

```python
df["Variable "]
```

```python

# Figure Size
fig = plt.figure(figsize=(10, 7))

# Horizontal Bar Plot
```

```python
import numpy as np
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people =list(df["Variable "])
y_pos = np.arange(len(people))
scaled_Importance = list(df["Value"])

ax.barh(y_pos, scaled_Importance, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Scaled Importance')
ax.set_ylabel('Variables')
ax.set_title('Variable Importance Plot')

plt.show()
```
